```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def draw(im):
    im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(14, 14))
    plt.axis('off')
    plt.imshow(im)
    plt.plot()

clusters = 6
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 1e-5)
attempts = 10
flags = cv2.KMEANS_RANDOM_CENTERS
```

```python
image = cv2.imread("test_3.jpg")
samples = np.float32(image.reshape((-1, 3)))

print("Исходное изображение:")
draw(image)
```

```python
compactness, labels, centers = cv2.kmeans(samples, clusters, None, criteria, attempts, flags)
centers = np.uint8(centers)
image_seg = centers[labels.flatten()].reshape(image.shape)
```

```python
print("Изображение после кластеризации:")
draw(image_seg)
status = cv2.imwrite('result_3.bmp', image_seg)
```
